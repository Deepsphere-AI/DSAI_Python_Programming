```python

# *********************************************************************************************************************
  
  # File Name 	:   CSLAB_GRAPHING_MATPLOTLIB_USING_DEFAULTS_ANNOTATING_POINTS_LEGENDS_V1
  # Purpose 	:   A Program in Python for Using Defaults & Annota in matplotlib
  # Author	:   Deepsphere.ai
  # Reviewer 	:   Jothi Periasamy
  # Date 	:   10/25/2022
  # Version	:   1.0	
  
# ***********************************************************************************************************************
   
## Program Description : Program for Using Defaults & Annota in matplotlib in Python

## Python Development Environment & Runtime - Python, Anaconda

from pylab import *

import numpy as vAR_np

vAR_CSLAB_X = vAR_np.linspace(-vAR_np.pi, vAR_np.pi, 256,endpoint=True)

vAR_CSLAB_C,vAR_CSLAB_S = vAR_np.cos(vAR_CSLAB_X), vAR_np.sin(vAR_CSLAB_X)

plot(vAR_CSLAB_X,vAR_CSLAB_C)
 
plot(vAR_CSLAB_X,vAR_CSLAB_S)

xlim(-4.0,4.0)

ylim(-1.0,1.0)

xlim(vAR_CSLAB_X.min()*1.1, vAR_CSLAB_X.max()*1.1)

ylim(vAR_CSLAB_C.min()*1.1, vAR_CSLAB_C.max()*1.1)

vAR_CSLAB_xmin, vAR_CSLAB_xmax = vAR_CSLAB_X.min(), vAR_CSLAB_X.max()

vAR_CSLAB_dx = (vAR_CSLAB_xmax - vAR_CSLAB_xmin) * 0.2

xlim(vAR_CSLAB_xmin - vAR_CSLAB_dx, vAR_CSLAB_xmax + vAR_CSLAB_dx)

xticks( [-vAR_np.pi, -vAR_np.pi/2, 0, vAR_np.pi/2, vAR_np.pi])

yticks([-1, 0, +1])

xticks([-vAR_np.pi, -vAR_np.pi/2, 0, vAR_np.pi/2, vAR_np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

vAR_CSLAB_ax = gca()

vAR_CSLAB_ax.spines['right'].set_color('none')

vAR_CSLAB_ax.spines['top'].set_color('none')

vAR_CSLAB_ax.xaxis.set_ticks_position('bottom')

vAR_CSLAB_ax.spines['bottom'].set_position(('data',0))
 
vAR_CSLAB_ax.yaxis.set_ticks_position('left')
 
vAR_CSLAB_ax.spines['left'].set_position(('data',0))

plot(vAR_CSLAB_X, vAR_CSLAB_C, color="blue", linewidth=2.5, linestyle="-", label="cosine")

plot(vAR_CSLAB_X, vAR_CSLAB_S, color="red",  linewidth=2.5, linestyle="-", label="sine")

legend(loc='upper left')

vAR_CSLAB_t = 2*vAR_np.pi/3

plot([vAR_CSLAB_t,vAR_CSLAB_t],[0,vAR_np.cos(vAR_CSLAB_t)], color ='blue', linewidth=2.5, linestyle="--")

scatter([vAR_CSLAB_t,],[vAR_np.cos(vAR_CSLAB_t),], 50, color ='blue')

annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(vAR_CSLAB_t, vAR_np.sin(vAR_CSLAB_t)), xycoords='data',xytext=(+10, +30), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plot([vAR_CSLAB_t,vAR_CSLAB_t],[0,vAR_np.sin(vAR_CSLAB_t)], color ='red', linewidth=2.5, linestyle="--")

scatter([vAR_CSLAB_t,],[vAR_np.sin(vAR_CSLAB_t),], 50, color ='red')
    
annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',xy=(vAR_CSLAB_t, vAR_np.cos(vAR_CSLAB_t)), xycoords='data',xytext=(-90, -50), textcoords='offset points', fontsize=16,arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
 
show()    
  

# ****************************************************************************************************************************
#   Disclaimer.

# We are providing this code block strictly for learning and researching, this is not a production
# ready code. We have no liability on this particular code under any circumstances; users should use
# this code on their own risk. All software, hardware and othr products that are referenced in these 
# materials belong to the respective vendor who developed or who owns this product.

# ****************************************************************************************************************************
  
```
